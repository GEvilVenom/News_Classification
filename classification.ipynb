```python
df=pd.read_csv("NEWSAPIDATA.csv")
```

```python
text=df['URL']
```

```python
preproc_text=[]
for t in text:
    preproc_text.append(t.split("/"))
```

```python
final_text=[]
for i in preproc_text:
    if i[-1]=="":
        i.remove(i[-1])
    final_text.append(i[-1])
```

```python
final_text=[f.replace("-"," ") for f in final_text]
final_text=[f.replace(".html","") for f in final_text]
```

```python
from gensim.parsing.preprocessing import remove_stopwords
sents=[]
for i in final_text:
    sents.append(remove_stopwords(i))
```

```python
df['clean_text']=sents
```
